# International Macroeconomics

**A dashboard of interactive models** — built for students of intermediate macroeconomics at UVA, free to reuse for any course.

Each chapter is a self-contained Jupyter notebook with a working model: clean math, an interactive simulator, and a short economic interpretation. You can read straight through, jump to a topic, or open any chapter live in your browser via the **🚀 launch** button at the top-right of every page.

```{admonition} New here? Start with the math primer.
:class: tip
If you haven't taken intermediate macro before, open the **[Math Primer](Preliminaries.ipynb)** first — growth rates, compounding, and the log approximation underpin almost every chapter that follows.
```

```{admonition} Run any chapter in your browser — no install required.
:class: important
Click [**Open the live JupyterLite workspace →**](../lite/lab/index.html) to launch every notebook in this book with a real Python kernel running entirely in your browser (via Pyodide). Sliders, plots, and `ipywidgets` all work; nothing is sent to a server.
```

## How to use this book

- **Read mode** — click any chapter in the sidebar. The previous / next buttons at the bottom of each page walk you through the canonical order. The sidebar is the source of truth — there is no parallel index page to keep in sync.
- **Run mode** — use the **🚀 launch** button in the top-right of any chapter to open it live (no install required) via JupyterLite, Binder, or Colab. Edit sliders and parameters, see plots update.
- **Download** — every chapter has a download icon for the `.ipynb` source so you can run it locally or fork it for an assignment.
- **Stuck?** — every page has an *Open issue* button in the top-right that files a GitHub issue against this book. Use it for typos, broken cells, or content questions.

```{admonition} Prerequisites
:class: note
Single-variable calculus, basic statistics, and an intro-macro course (so terms like *aggregate demand*, *real GDP*, and *steady state* are familiar). Python is **not** required to *read* the book — only to modify the simulators.
```

_Originally built for an intermediate macroeconomics course at the University of Virginia. Maintained as a living teaching resource — pull requests welcome._
